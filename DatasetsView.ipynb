```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=3, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

# i did it 3 rows and 2 columns

aapl[['Close']].plot(ax=axes[0,0])
axes[0,0].set_title('APPLE')

goog[['Close']].plot(ax=axes[0,1])
axes[0,1].set_title('GOOGLE')

msft[['Close']].plot(ax=axes[1,0])
axes[1,0].set_title('MICROSOFT')

amzn[['Close']].plot(ax=axes[1,1])
axes[1,1].set_title('AMAZON')

nbk[['Close']].plot(ax=axes[2,0])
axes[2,0].set_title('NATIONAL BANK OF KUWAIT')

csco[['Close']].plot(ax=axes[2,1])
axes[2,1].set_title('CISCO')

fig.tight_layout()
```
