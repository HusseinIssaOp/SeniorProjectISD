```python
#explain the moving averages


from google.colab import files
import pandas as pd

uploaded = files.upload()
x = list(uploaded.keys())[0]
AAPL = pd.read_csv(x)


uploaded = files.upload()
y = list(uploaded.keys())[0]
GOOG = pd.read_csv(y)


uploaded = files.upload()
z = list(uploaded.keys())[0]
MSFT = pd.read_csv(z)


uploaded = files.upload()
z = list(uploaded.keys())[0]
AMZN = pd.read_csv(z)
AMZN
```

```python
company_list = [AAPL, GOOG, MSFT, AMZN]
tech_list = ['AAPL', 'GOOG', 'MSFT', 'AMZN']
```

```python
AAPL.info()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Close'].plot()
    plt.ylabel('Close')
    plt.xlabel('time')
    plt.title(f"Closing Price of {tech_list[i - 1]}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Volume'].plot()
    plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title(f"Sales Volume for {tech_list[i - 1]}")

plt.tight_layout()
```

```python
ma_day = [10, 20, 50]

for ma in ma_day:
    for company in company_list:
        column_name = f"MA for {ma} days"
        company[column_name] = company['Close'].rolling(ma).mean()


fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

AAPL[['Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,0])
axes[0,0].set_title('APPLE')

GOOG[['Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,1])
axes[0,1].set_title('GOOGLE')

MSFT[['Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,0])
axes[1,0].set_title('MICROSOFT')

AMZN[['Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,1])
axes[1,1].set_title('AMAZON')

fig.tight_layout()
```

```python
ma_day = [10, 20, 50]

for ma in ma_day:
    column_name = f"MA for {ma} days"
    AMZN[column_name] = AMZN['Close'].rolling(ma).mean()

last_100_days = AMZN.iloc[-100:]

last_100_days[['Date', 'Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(x='Date', figsize=(10, 6))
plt.title('Amzn Moving Averages (Last 100 Days)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
